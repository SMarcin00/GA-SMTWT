```python
import pandas as pd
ga_df = pd.read_csv("data/ga_population.csv")
ga_df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

mean_scores = ga_df.groupby("PopulationSize").BestIndividualScore.mean()
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

mean_scores.plot(ax=ax, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)

ax.set_title(f"Average Fitness Score of Genetic Algorithm\n", fontsize=16, fontweight='bold')
ax.text(0.5, 1.15, f"TaskNumber: {ga_df['TaskNumber'][0]}\ntSize: {ga_df['tSize'][0]}\nMutationProbability: {ga_df['MutationProbability'][0]}\nIterations: {ga_df['Iterations'][0]}", fontsize=10, ha='center', transform=ax.transAxes)
ax.set_ylabel("Fitness Score", fontsize=12)
ax.set_xlabel("Population Size", fontsize=12)

ax.tick_params(axis='both', which='major', labelsize=10, length=6, width=1)
ax.tick_params(axis='both', which='minor', length=4, width=1)
ax.grid(alpha=0.3)
sns.despine(ax=ax, top=True, right=True)
plt.tight_layout()
plt.show()
```
